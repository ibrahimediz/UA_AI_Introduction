## Series

```python
liste = ["Elma","Portakal","Limon"]
fiyat = [20,7,19]
print(liste)
print(fiyat)
```

```python
seri1 = pd.Series(liste)
seri1
```

```python
seri2 = pd.Series(fiyat)
seri2
```

```python
seri3 = pd.Series(fiyat,index=["a","b","c"])
seri3
```

## DataFrame

```python
df = pd.DataFrame({"İsim":seri1,"Fiyat":seri2,"Miktar":[20,40,30]})
df
```

```python
df = pd.DataFrame({"İsim":seri1,"Fiyat":seri3})
df
```
